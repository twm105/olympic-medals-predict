```python
import pandas as pd
```

```python
data_dict = {'name': ['Andy', 'Bob', 'Carrie'],
             'matts1': [1, 2, 3],
             'matts2': [2, 2, 2],
             'matts3': [3, 2, 1],
             'matts4': [1, 3, 3],
             'matts5': [3, 2, 3],
             'dip': [1, 2, 2],
             'tet': [3, 2, 3],
             'pol': [1, 1, 3],
             'typ': [1, 2, 1],
             'hep': [2, 2, 3]}

df_in = pd.DataFrame(data_dict)
df_in.head()
```

```python
df = df_in.melt(id_vars=['name'],
                var_name='factor')
df.head()
```

```python
df['group'] = 'vax'
df.loc[df.factor.str.contains('matts'), 'group'] = 'matts'
df.head()
```

```python
pd.pivot_table(data=df, 
               index=['group', 'factor'],
               columns='value',
               values='name',
               aggfunc='count',
               fill_value=0)
```
